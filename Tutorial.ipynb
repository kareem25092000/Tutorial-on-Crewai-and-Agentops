# importing libraries and Connecting to agentops 

```python
import os
from crewai import Task ,LLM ,Crew ,Process ,Agent
from dotenv import load_dotenv
import agentops

load_dotenv()


AGENTOPS_API_KEY = os.getenv("AGENTOPS_API_KEY")
agentops.init(
    api_key=AGENTOPS_API_KEY,
    default_tags=['crewai']
)
```

# implementing the first crew

- we will use gemini-3-flash-preview so make sure that the api key in `.env` file with name `GEMINI_API_KEY` <br/><br/>
- we will create two agents :<br/><br/>
1 - the first one is an egyptian author with 10 years of experience that is inspired by dostoevsky and his goal is to create stories given some details about it , and in this project the task for this agent is to Generate story with specific theme and specific name for the main character with specific subject<br/>
2 -the second agent is German writer with 5 years of experience and his task is to summarize the story generated from the first agent while keeping the style .

```python
llm = LLM(model="gemini/gemini-3-flash-preview",
          api_key = os.getenv("GEMINI_API_KEY"),
          temperature=0)
```

```python
# the first agent

author_agent = Agent(
    role = "Author",
    goal = "to create stories given some details about it",
    backstory = ("Egyptian writer with 10 years of experience that is "
                 "inspired by dostoevsky"),
    llm = llm,
    verbose = True
)

# the task for the first agent
author_task = Task(
    description = ("Generate story that has the theme {theme} , the main character name "
                   "is {name} and the subject is {subject}"),
    expected_output = "a short story that has maximum 500 words",
    agent = author_agent
)
```

```python
# the second agent
summarizer_agent = Agent(
    role = "Summarize",
    goal = "to summarize stories",
    backstory = ("German writer with 5 years of experience"),
    llm = llm,
    verbose = True
)

# the task for the second agent
summarize_task = Task(
    description = "to summarize stories but kepping the style",
    expected_output = "a summarized version of the given story with the main ideas kept",
    agent = summarizer_agent
)
```

```python
# the first crew
first_crew = Crew(
    agents=[
        author_agent, 
        summarizer_agent
    ],
    tasks=[
        author_task,
        summarize_task
    ],
    process=Process.sequential
)
```

```python
# calling the first crew

results = first_crew.kickoff(inputs={
    "theme" : "horror",
    "name" : "ahmed" ,
    "subject" : "hauneted house with ghosts"
})
```

# implementing the second crew

- we will use gemini-3-flash-preview so make sure that the api key in `.env` file with name `GEMINI_API_KEY` <br/><br/>
- we will create four agents :<br/><br/>
1 - Writer Agent: Produces a long-form, well-structured article explaining the project in detail.<br/>
2 - Summarizer Agent: Generates short summaries, bullet points, and TL;DR sections.<br/>
3 - Fact-Checker Agent: Verifies technical accuracy of the content.<br/>
4 - Metadata Agent: Creates SEO-friendly titles, tags, and structured metadata for publishing.

```python
llm = LLM(model="gemini/gemini-3-flash-preview",
          api_key = os.getenv("GEMINI_API_KEY"),
          temperature=0)
```

```python
writer_agent = Agent(
    role = "writer",
    goal = ("Produces a long-form, well-structured article explaining "
    "a specific project in detail"),
    backstory = ("egyptian programmer with 10 years of experience"),
    llm = llm,
    verbose = True
)

writer_task = Task(
    description = ("Generate well structured article given the name of a topic :"
                    "{topic}"),
    expected_output = "a well structured detailed article",
    agent = writer_agent
)
```

```python
summarizer_agent = Agent(
    role = "summarizer",
    goal = "to summarize an article while keeping the important information",
    backstory = ("german programmer with 5 years of experience"),
    llm = llm,
    verbose = True
)

summarizer_task = Task(
    description = ("summarize an article without removing any important details"),
    expected_output = "a smaller artcicle with half the words of the original",
    agent = summarizer_agent
)
```

```python
fact_checker_agent = Agent(
    role = "fact checker",
    goal = "to change any misinformation in an article",
    backstory = ("american programmer that has 20 years of experience"),
    llm = llm,
    verbose = True
)

fact_checker_task = Task(
    description = ("Check for validity of an article and fix any wrong information"),
    expected_output = "a correct version of the input article",
    agent = fact_checker_agent
)
```

```python
metadata_agent = Agent(
    role = "metadata agent",
    goal = "Creates SEO-friendly titles, tags, and structured metadata for publishing",
    backstory = ("Egyptian writer with 10 years of experience"),
    llm = llm,
    verbose = True
)

metadata_task = Task(
    description = ("extract the meta data for an article to make searching for it "
                    "easier on the internet"),
    expected_output = "titles ,tags and metadata for input article",
    agent = metadata_agent
)
```

```python
# the second crew
second_crew = Crew(
    agents=[
        writer_agent, 
        summarizer_agent,
        fact_checker_agent,
        metadata_agent
    ],
    tasks=[
        writer_task,
        summarizer_task,
        fact_checker_task,
        metadata_task
    ],
    process=Process.sequential
)
```

```python
# calling the second crew

results = second_crew.kickoff(inputs={
    "topic" : "functional programming"
})
```
